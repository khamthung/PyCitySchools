```python
import pandas as pd
import numpy as np
```

```python
school_data = pd.read_csv('Resources/schools_complete.csv')
student_data = pd.read_csv('Resources/students_complete.csv')
```

```python
df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
df.head(5)
```

```python
total_schools = len(df["School ID"].unique())
total_students = len(df["Student ID"].unique())
total_budget = '${:,.2f}'.format(school_data["budget"].sum())
average_math_score = df["math_score"].mean()
average_reading_score = df["reading_score"].mean()
percent_passing_math = (len(df.loc[df["math_score"] >= 70]["Student ID"].unique())/total_students)*100
percent_passing_reading = (len(df.loc[df["reading_score"] >= 70]["Student ID"].unique())/total_students)*100
passing_overall = (average_math_score+average_reading_score)/2
```

```python
district_summary = pd.DataFrame({"Total Schools": total_schools,
                              "Total Students": total_students,
                              "Total Budget": total_budget,
                              "Average Math Score": average_math_score,
                              "Average Reading Score": average_reading_score,
                              "% Passing Math": percent_passing_math,
                              "% Passing Reading": percent_passing_reading,
                              "% Overall Passing Rate": passing_overall
                              },index=["District Summary"])
district_summary
```

```python
school_type_groupby_school = df.groupby(['school_name'])["type"].unique()
total_students_groupby_school = df.groupby(['school_name'])["Student ID"].count()
total_budget_groupby_school = df.groupby(['school_name'])["budget"].unique()
budget_per_student = total_budget_groupby_school/total_students_groupby_school
average_math_score_groupby_school = df.groupby(['school_name'])["math_score"].mean()
average_reading_score_groupby_school = df.groupby(['school_name'])["reading_score"].mean()
percent_passing_math_groupby_school = df[df["math_score"] >= 70].groupby(['school_name'])["math_score"].count()/total_students_groupby_school*100
percent_passing_reading_groupby_school = df[df["reading_score"] >= 70].groupby(['school_name'])["reading_score"].count()/total_students_groupby_school*100
passing_overall_groupby_school = (percent_passing_math_groupby_school+percent_passing_reading_groupby_school)/2
```

```python
school_summary = pd.DataFrame({"School Type": school_type_groupby_school,
                              "Total Students": total_students_groupby_school,
                              "Total Budget": total_budget_groupby_school,
                              "Budget Per Student": budget_per_student,
                              "Average Math Score": average_math_score_groupby_school,
                              "Average Reading Score": average_reading_score_groupby_school,
                              "% Passing Math": percent_passing_math_groupby_school,
                              "% Passing Reading": percent_passing_reading_groupby_school,
                              "% Overall Passing Rate": passing_overall_groupby_school
                              })
school_summary = school_summary.sort_values("% Overall Passing Rate", ascending=False)
school_summary.head(5)
```

```python
school_summary = school_summary.sort_values("% Overall Passing Rate", ascending=True)
school_summary.head(5)
```

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
df["budget_per_student"] = df["budget"]/df["size"]
df["spending_ranges"] = pd.cut(df["budget_per_student"], bins=spending_bins, labels=group_names)
df
```

```python
total_students_groupby_spending_ranges = df.groupby(['spending_ranges'])["Student ID"].count()
average_math_score_groupby_spending_ranges = df.groupby(['spending_ranges'])["math_score"].mean()
average_reading_score_groupby_spending_ranges = df.groupby(['spending_ranges'])["reading_score"].mean()
percent_passing_math_groupby_spending_ranges = df[df["math_score"] >= 70].groupby(['spending_ranges'])["math_score"].count()/total_students_groupby_spending_ranges*100
percent_passing_reading_groupby_spending_ranges = df[df["reading_score"] >= 70].groupby(['spending_ranges'])["reading_score"].count()/total_students_groupby_spending_ranges*100
passing_overall_groupby_spending_ranges = (percent_passing_math_groupby_spending_ranges+percent_passing_reading_groupby_spending_ranges)/2
```

```python
spending_ranges_summary = pd.DataFrame({
                              "Average Math Score": average_math_score_groupby_spending_ranges,
                              "Average Reading Score": average_reading_score_groupby_spending_ranges,
                              "% Passing Math": percent_passing_math_groupby_spending_ranges,
                              "% Passing Reading": percent_passing_reading_groupby_spending_ranges,
                              "% Overall Passing Rate": passing_overall_groupby_spending_ranges
                              })
spending_ranges_summary.head(5)
```
